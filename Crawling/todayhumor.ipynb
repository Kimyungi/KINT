```python
def canfetch(url, agent='*', path='/'):
    robot = RobotFileParser(urljoin(url, '/robots.txt'))
    robot.read()
    return robot.can_fetch(agent, urlparse(url)[2])
    
def download(url, params={}, headers={}, method='GET', limit=3):
    method = method.upper()
    if canfetch(url) == False:
        print('[Error] ' + url)
#     else: # 실제 수집할 때, 제약사항이 많으므로 여기선 잠시 해제
    try:
        resp = request(method, url,
               params=params if method=='GET' else {},
               data=params if method=='POST' else {},
               headers=headers)
        resp.raise_for_status()
    except HTTPError as e:
        if limit > 0 and e.response.status_code >= 500:
            print(limit)
            time.sleep(1) # => random
            resp = download(url, params, headers, method, limit-1)
        else:
            print('[{}] '.format(e.response.status_code) + url)
            print(e.response.status_code)
            print(e.response.reason)
            print(e.response.headers)
    return resp
```

```python
from requests.compat import urlparse, urljoin
import sqlite3
from requests import Session
from bs4 import BeautifulSoup

conn = sqlite3.connect('todayhumor1.db')
cur = conn.cursor()
cur.execute('''
    CREATE TABLE todayhumor(
        pk INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        url TEXT NOT NULL,
        head TEXT NOT NULL,
        body TEXT NOT NULL,
        comment TEXT NOT NULL,
        date TEXT NOT NULL
    );
''')
conn.commit()

url = 'http://www.todayhumor.co.kr/board/list.php?kind=total'
session = Session()

urls = list()
seen = list()

urls.append(url)

count = 0

while urls:
    count += 1
    seed = urls.pop(0)

    resp = session.request('GET', seed)
    dom = BeautifulSoup(resp.text, 'lxml')

    seen.append(seed)

    for _ in [_['href'] for _ in dom.select('td.subject > a, td > a ') 
              if _.has_attr('href')]:
        newUrls = urljoin(seed, _)
        if newUrls not in urls \
                and newUrls not in seen:
            urls.append(newUrls)
    if dom.select_one('#containerInner > div.viewSubjectDiv > div') != None :
        head = str(dom.select_one('#containerInner > div.viewSubjectDiv > div').text.strip()),
        body = str(dom.select_one('#containerInner > div.contentContainer > div.viewContent').text.strip()),
        comments = str([_.text.strip() for _ in dom.select('memoWrapperDiv div.memoContent')]),
        date = dom.select_one('#containerInner > div.writerInfoContainer > div.writerInfoContents > div:nth-child(7)').string
        cur.execute('''
            INSERT INTO todayhumor(url, head, body, comment, date) VALUES(?,?,?,?,?)
        ''', [seed, head, body, comments, date])
        conn.commit()

    if count % 100 == 0:
        print(count)
```
