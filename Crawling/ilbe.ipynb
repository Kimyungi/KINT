```python
from requests.compat import urlparse, urljoin
from requests.exceptions import HTTPError
import time
import sqlite3
from requests import Session
from bs4 import BeautifulSoup
import re
```

```python
conn = sqlite3.connect('ilbe.db')
cur = conn.cursor()
cur.executescript('''
    DROP TABLE IF EXISTS ilbe;
    CREATE TABLE ilbe(
        pk INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        url TEXT NOT NULL,
        head TEXT NOT NULL,
        body TEXT NOT NULL,
        comment TEXT NOT NULL,
        date TEXT NOT NULL
    );
''')
```

```python
url = 'https://www.ilbe.com/list/free'
params = {
    'page':'',
    'listStyle':'list',
}
page = list()
page.append(1)
params['page'] = page.pop(0)

session = Session()

urls = list()
seen = list()

urls.append(url)

count = 0

while urls:
    try:
        seed = urls.pop(0)

        resp = session.request('GET',seed, params=params)
        dom = BeautifulSoup(resp.text, 'lxml')
        time.sleep(0.02)

        seen.append(seed)

        if '/list' in urlparse(seed)[2]:
            for _ in [_.text.strip() for _ in dom.select('.paginate > a')]:
                if _ == '다음':
                    for i in list(range(10*count+1,10*count+10)):
                        page.append(i)
                    count += 1
            for _ in [_['href'] for _ in dom.select('.board-body > li > .title > a')
                      if _.has_attr('href') and not _.has_attr('style')\
                      and re.match(r'(?:https?:/)?/\w+(?:[./]\w+)+', _['href'])]:
                newUrls = urljoin(seed, _)
                if newUrls not in urls\
                and newUrls not in seen:
                    urls.append(newUrls)
        if dom.select_one('.post-content') != None:
            head = dom.select_one('.post-header > h3 > a').text.strip()
            body = dom.select_one('.post-content').text.strip()
            comments = ''.join([_.text.strip() for _ in dom.select('#comment_wrap_in .comment-box > .cmt')])
            date = dom.select_one('.date').text.strip()
            cur.execute('''
                INSERT INTO ilbe(url, head, body, comment, date) VALUES(?,?,?,?,?)
            ''', [seed, head, body, comments, date])
            conn.commit()
        if not urls:
            urls.append(url)
            params['page'] = page.pop(0)
        
        if count%100 == 0:
            print(count)

    except:
        print('Error')
```
